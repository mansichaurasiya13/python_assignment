```python
#3. Compare and contrast mutable and immutable objects in Python with examples.
#Ans:- In Python, objects can be classified as mutable or immutable based on their ability to change after creation.

#Mutable Objects:

#- Can be changed after creation
#- Modifications affect the original object
#- Examples: lists, dictionaries, sets, user-defined classes

#Example:

#my_list = [1, 2, 3]
#my_list.append(4)
#print(my_list)  # Output: [1, 2, 3, 4]


#Immutable Objects:

#- Cannot be changed after creation
#- Modifications create a new object
#- Examples: integers, floats, strings, tuples

#Example:

#my_string = "hello"
#my_string += " world"
#print(my_string)  # Output: "hello world"
#print(id(my_string))  # Different ID, new object created


#Key differences:

#- Mutable objects can be modified in-place, while immutable objects require creating a new object for modifications.
#- Mutable objects are generally more memory-efficient, as they don't require creating new objects for each modification.
#- Immutable objects are safer for concurrent access and can be used as dictionary keys, while mutable objects cannot.

#When to use each:

#- Use mutable objects when you need to modify data frequently, such as in algorithms or data processing.
#- Use immutable objects when you need to ensure data integrity, thread safety, or use objects as dictionary keys.
```

```python
#5. Explain the concept of type casting inPython with examples.
#Ans:- Type casting in Python refers to the process of converting a value from one data type to another. This can be done using built-in functions or methods. Here are some examples:

#1. Integers to Floats:


#x = 5
#y = float(x)
#print(y)  # Output: 5.0


#1. Floats to Integers:


#x = 5.5
#y = int(x)
#print(y)  # Output: 5


#1. Strings to Integers or Floats:


#x = "5"
#y = int(x)
#print(y)  # Output: 5

#x = "5.5"
#y = float(x)
#print(y)  # Output: 5.5


#1. Integers or Floats to Strings:

#x = 5
#y = str(x)
#print(y)  # Output: "5"

#x = 5.5
#y = str(x)
#print(y)  # Output: "5.5"


#1. Boolean to Integer:


#x = True
#y = int(x)
#print(y)  # Output: 1

#x = False
#y = int(x)
#print(y)  # Output: 0

#Note that when casting a float to an integer, the decimal part is truncated, not rounded.

#Type casting is useful when you need to perform operations that require a specific data type, or when working with external data sources that provide data in a different format than what your program expects.
```
